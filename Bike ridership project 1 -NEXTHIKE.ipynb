### Finding outliers by IQR method

```python
# DEFINING FUNCTION TO FIND OUTLIERS OF ALL COLUMNS IN FINAL DATASET

FD_col=['hr', 'temp', 'atemp', 'hum', 'windspeed', 'registered', 'cnt']

def outliers(df,FD_col):
    a = 0  # Initialize 'a' before the loop
    while a < len(FD_col):
        col = FD_col[a]  # Define 'col' before using it
        print(FD_col[a])
        
        # DETECTING OUTLIERS
        
        q1=df[col].quantile(.25) # computing first quartile
        print('FIRST QUARTILE: ','\n',q1)
        
        q3=df[col].quantile(.75) # computing third quartile
        print('THIRD QUATILE: ','\n',q3)
        
        IQR=q3-q1   # computing IQR
        print('INTER QUARTILE RANGE : ','\n',IQR)
    
        upper = q3 +1.5*IQR    # computing upper bound
        print('UPPER RANGE: ','\n',upper)
    
        lower = q1 - 1.5*IQR   # computing lower bound
        print('LOWER RANGE: ','\n',lower)
    
        # REMOVING OUTLIERS
    
        outliers_removed=df.index[(df[col] < lower ) | (df[col] > upper )]
    
        removed_list=outliers_removed.tolist() # adding the outliers to a list
    
        print(removed_list) # printing the list of outliers removed  
        a += 1
```

```python
outliers(final_dataset,FD_col)
```
